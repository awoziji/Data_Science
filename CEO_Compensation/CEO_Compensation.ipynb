## Creating a Mathematical Model
We are also interested to see if we can create a mathematical model that will help us predict the CEO compensation based on company size, i.e., the company ranking. 

Because of my chemistry background, the second I ploted the data set it made me think of an exponential function. In particular, it reminded me of the __[Arrenhius equation](https://en.wikipedia.org/wiki/Arrhenius_equation)__: $$k = A\exp\left(\frac{-E_a}{RT}\right),$$ where $k$ is the rate constant, $A$ is called the pre-exponential factor (the frequency of collisions in the correct orientation for the reaction to occur), $E_a$ is the activation energy, $R$ is a constant and $T$ is the temperature.

Now we want to construct an equation just to give us a view as to what this would look like. The first thing we can consider is the floor of the function. The minimum value is going to need to match the compensation of the lowest paid CEO on our list of 500. We will use `ceo_pay_by_year.Compensation.iloc[-1]` to determine the minimum value. To determine the value at a ranking of one we need to set the equation so that it has a value matching `ceo_pay_by_year.Compensation.iloc[0]`. We will now try to find the approximate values for our equation, which has either of the following forms: $$\begin{align}c &= \alpha\exp(-\beta\times\mathrm{rank})+b & (A)\\
c &= \gamma\exp\left(\frac{\delta}{\mathrm{rank}}\right)+b & (B)\end{align}$$
where $c$ is the compensation and $b$ is the minimum compensation or base level in the data set.

````python
# We're using the floor to prevent having to calculate log(0)
base = floor(ceo_pay_by_year.Compensation.min())
top = ceo_pay_by_year.Compensation.iloc[0]
rank1 = ceo_pay_by_year.Rank.iloc[0]
rank500 = ceo_pay_by_year.Rank.iloc[-1]

# Determine values for alpha, beta, gamma and delta using two systems of equations
beta = - (log(top-base)) / (rank1 - rank500)
alpha = (top - base) * np.exp(beta * rank1)

delta =  (log(top-base)) / ( (1/rank1) - (1/rank500))
gamma = (top - base) * np.exp(-delta * rank1)

print("Top Pay: {:.3f}\tMin Pay: {}\nAlpha: {:.6f}\tBeta: {:.6f}\n\
Gamma: {:.6f}\t\tDelta: {:.6f}".format(top,base,alpha,beta,gamma,delta))

print('\n\n\t\tModel A:\tModel B:\nRank 1:\t\t{:f}\t{:f}\n\
Rank 500:\t{:f}\t{:f}\n'.format(alpha*np.exp(-beta*rank1)+base,
                                gamma*np.exp(delta/rank1)+base,
                                alpha*np.exp(-beta*rank500)+base,
                                gamma*np.exp(delta/rank500)+base))
````

## Plotting the Models

````python
fig = plt.figure(figsize=(15,4))
plt.subplot(1, 3, 1)
plt.plot(
    np.arange(1,500,10),
    list(map(lambda rank: alpha*np.exp(-beta*rank)+base,
    np.arange(1,500,10))),
    '-o', c='xkcd:spruce', alpha=0.7 )
plt.ylabel('Compensation (USD)')
plt.xlabel('Rank')
plt.title('Model A')

plt.subplot(1, 3, 2)
# I can't for the life of me remember where I derived
# this from which prompted me to come up with the above ones.
factor = 0.04678 
plt.plot(
    np.arange(1,500,10),
    list(map(lambda x: ((top)/np.exp(-factor))*np.exp(-x*factor) + base,
    np.arange(1,500,10))),
    '-o', c='xkcd:dark grass green', alpha=0.7)
plt.ylabel('Compensation (USD)')
plt.xlabel('Rank')
plt.title('Model A Variant')

plt.subplot(1, 3, 3)
plt.plot(
    np.arange(1,500,10),
    list(map(lambda rank: gamma*np.exp(delta/rank)+base,
    np.arange(1,500,10))),
    '-o', c='xkcd:grey teal', alpha=0.7 )
plt.ylabel('Compensation (USD)')
plt.xlabel('Rank')
plt.title('Model B')
````

The Arrenhius equation can be linearized by taking the $\log$ of both sides of the equation (this is how it is done in chemistry labs to simplify calcuations or find the activation energy). The new equation looks like this: $$\ln(k) = \frac{-E_a}{R}\left(\frac{1}{T}\right)+\ln(A).$$ When we translate this to our two variations (A) and (B) we get:
$$\begin{align}
\log(c-b) &= -\beta R +  \log \alpha & (A)\\
\log(c-b) &= \frac{\delta}{R} +  \log \gamma & (B)
\end{align}$$

## Plotting the Linearized Models

````python
fig = plt.figure(figsize=(17,5))
plt.subplot(1, 3, 1)
plt.plot(
    np.arange(1,500,10),
    list(map(lambda rank: -beta*rank + log(alpha),
    np.arange(1,500,10))),
    '-o', c='xkcd:spruce', alpha=0.7 )
plt.ylabel('log(Compensation - Base)')
plt.xlabel('Rank')
plt.title('Model A')

plt.subplot(1, 3, 2)
# I can't for the life of me remember where I derived
# this from which prompted me to come up with the above ones.
factor = 0.04678 
plt.plot(
    np.arange(1,500,10),
    list(map(lambda x: log(top/np.exp(-factor))-factor*x,
    np.arange(1,500,10))),
    '-o', c='xkcd:dark grass green', alpha=0.7)
plt.ylabel('log(Compensation - Base)')
plt.xlabel('Rank')
plt.title('Model A Variant')

plt.subplot(1, 3, 3)
plt.plot(
    np.arange(1,500,10),
    list(map(lambda rank: delta / rank + log(gamma),
    np.arange(1,500,10))),
    '-o', c='xkcd:grey teal', alpha=0.7 )
plt.ylabel('log(Compensation - Base)')
plt.xlabel('Rank')
plt.title('Model B')
````

From the above graphs of the linearized versions we can see that we have to discard Model B because it's not actually linear in this form.

## Segment the data by Gender
Since we are tyring to see if CEO pay is a function of company rank and avoids gender, we will separate the data by gender before building a model and use the female CEO data as a holdout test. Then we'll randomly sample the male data to also make a test set.

````python
ceo_pay_by_year_m = ceo_pay_by_year[ceo_pay_by_year['Ceo'].\
                            isin(list(male_ceo_pay['Ceo'].values))].copy()
ceo_pay_by_year_f = ceo_pay_by_year[ceo_pay_by_year['Ceo'].\
                            isin(list(female_ceo_pay['Ceo'].values))].copy()

ceo_pay_by_year_m['log(Compensation)'] = ceo_pay_by_year_m['Compensation'].\
                                            apply(lambda s: log(s - base))
ceo_pay_by_year_f['log(Compensation)'] = ceo_pay_by_year_f['Compensation'].\
                                            apply(lambda s: log(s - base))

ceo_pay_2015_m = ceo_pay_by_year_m[ceo_pay_by_year_m.Year == 2015].copy()
ceo_pay_2016_m = ceo_pay_by_year_m[ceo_pay_by_year_m.Year == 2016].copy()
ceo_pay_2017_m = ceo_pay_by_year_m[ceo_pay_by_year_m.Year == 2017].copy()

ceo_pay_2015_f = ceo_pay_by_year_f[ceo_pay_by_year_f.Year == 2015].copy()
ceo_pay_2016_f = ceo_pay_by_year_f[ceo_pay_by_year_f.Year == 2016].copy()
ceo_pay_2017_f = ceo_pay_by_year_f[ceo_pay_by_year_f.Year == 2017].copy()
````

## Split the data into training and test data

````python
X_train, X_test, y_train, y_test = train_test_split(
    ceo_pay_by_year_m['Rank'],
    ceo_pay_by_year_m['log(Compensation)'],
    test_size=0.33,
    random_state=42)
````

## Linear Models

````python
cv_test_errors_alpha = []
cv_test_errors_epsilon = []

epsilons = np.arange(1.0, 20.0, 0.05)
alphas = np.linspace(0.0001, 100.0)


for epsilon in epsilons:
    huber_model = HuberRegressor(epsilon=epsilon, alpha=0.0001)
    cv_test_error = -cross_val_score(
        huber_model,
        X_random_order.reshape(-1,1),
        y_random_order.ravel(),
        cv=5,  # Number of folds
        scoring='neg_mean_squared_error'
    )
    cv_test_errors_epsilon.append(cv_test_error.mean())


epsilon = epsilons[np.argmin(cv_test_errors_epsilon)]
for alpha in alphas:
    huber_model = HuberRegressor(epsilon=epsilon, alpha=alpha)
    cv_test_error = -cross_val_score(
        huber_model,
        X_random_order.reshape(-1,1),
        y_random_order.ravel(),
        cv=5,  # Number of folds
        scoring='neg_mean_squared_error'
    )
    cv_test_errors_alpha.append(cv_test_error.mean())
````

So the model seems to fit the data fairly well. Let's see if we can do better.

Let's check up to a `max_depth` of 15 and use MSE as a metric.

````python
max_depths = range(1, 16)
in_sample_errors = []
for max_depth in max_depths:
    y_pred = DecisionTreeRegressor(
        max_depth=max_depth).fit(X_train.values.reshape(-1,1),
        y_train.values.reshape(-1,1)).predict(X_train.values.reshape(-1,1))
    
    in_sample_errors.append(mean_squared_error(y_train.values.reshape(-1,1), y_pred))

test_errors = []
for max_depth in max_depths:
    est = DecisionTreeRegressor(max_depth=max_depth).fit(
        X_train.values.reshape(-1,1), 
        y_train.values.reshape(-1,1))
    y_pred = est.predict(X_test.values.reshape(-1,1))
    
    test_errors.append(mean_squared_error(y_test.values.reshape(-1,1), y_pred))

    
cv_test_errors = []
indices = np.random.permutation(range(ceo_pay_by_year_m['log(Compensation)'].size))
X_random_order, y_random_order = ceo_pay_by_year_m['Rank'].as_matrix().\
                                    astype(np.float)[indices],\
                                 ceo_pay_by_year_m['log(Compensation)'].as_matrix().\
                                    astype(np.float)[indices]

for max_depth in max_depths:
    est = DecisionTreeRegressor(max_depth=max_depth)
    cv_test_error = -cross_val_score(
        est,
        X_random_order.reshape(-1,1),
        y_random_order.reshape(-1,1),
        cv=5,  # Number of folds
        scoring='neg_mean_squared_error'
    )
    cv_test_errors.append(cv_test_error.mean())
    
plt.figure(figsize=(8,8))
plt.plot(max_depths, in_sample_errors, '-o', label='In-Sample Error', c='xkcd:cerulean')
plt.plot(max_depths, test_errors, '-o', label='Out-of-Sample Error', c='green')
plt.plot(max_depths, cv_test_errors, '-o', label='Cross-Validation Error', c='orange')
plt.xlabel('max_depth')
plt.ylabel('MSE')
plt.legend(loc='upper right')
````

It's hard to tell exactly which depth is optimal from the graph, so we'll use `np.argmin` to choose the lowest value from the `cv_test_errors`.

````python
opt_tree_model = DecisionTreeRegressor(max_depth=np.argmin(cv_test_errors))
opt_tree_model.fit(X = X_train.values.reshape(-1,1), y = y_train.values.reshape(-1,1))

print(opt_tree_model.get_params)
dot_data = StringIO() 
export_graphviz(opt_tree_model, out_file=dot_data) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
display.Image(graph.create_png())
````
